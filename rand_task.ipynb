```python
import random
import matplotlib.pyplot as pyplot
%matplotlib inline
import matplotlib.pylab as pylab

def get_rand_4():
    """
    Returns 1 or 2 or 3 or 4 with equal probability
    """
    return random.randrange(1, 5)

def get_rand_6():
    """
    Returns 1 or 2 or 3 or 4 or 5 or 6 with equal probability
    """
    result = '{0:02b}'.format(get_rand_4() - 1)[1:] + '{0:02b}'.format(get_rand_4() - 1)[1:] + '{0:02b}'.format(get_rand_4() - 1)[1:]
    return int(result, 2)
```

```python
DISTRIBUTION = dict()

for index in range(10000):
    random_6 = get_rand_6()
    if random_6 not in DISTRIBUTION.keys():
        DISTRIBUTION[random_6] = 0
    DISTRIBUTION[random_6] += 1

print DISTRIBUTION
```

```python
pylab.rcParams['figure.figsize'] = 16, 12

pyplot.xlabel('Random number')
pyplot.ylabel('Trials having this number')
pyplot.bar([x for (x, y) in DISTRIBUTION.items()], [y for (x, y) in DISTRIBUTION.items()])
pyplot.legend()

pyplot.show()
```
