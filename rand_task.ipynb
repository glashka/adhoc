```python
import random
import math
import matplotlib.pyplot as pyplot
%matplotlib inline
import matplotlib.pylab as pylab

def get_rand_4():
    """
    Returns 1 or 2 or 3 or 4 with equal probability
    """
    return random.randrange(1, 5)

def get_rand_6():
    """
    Returns 1 or 2 or 3 or 4 or 5 or 6 with equal probability
    """
    number = 0
    while number == 0:
        number = get_rand_4() - 1
    if get_rand_4() > 2:
        return number + 3
    return number

def get_rand_m(m):
    """
    Returns int in range(1,m+1) with equal probability
    """
    number_trials = int(math.ceil(math.log(m, 4)))
    result_binary = ''
    while result_binary == '' or int(result_binary, 2) > m - 1:
        result_binary = ''
        for i in range(number_trials):
            result_binary += '{0:02b}'.format(get_rand_4() - 1)
    return int(result_binary, 2) + 1
```

```python
DISTRIBUTION = dict()

for index in range(10000):
    random_6 = get_rand_6()
    if random_6 not in DISTRIBUTION.keys():
        DISTRIBUTION[random_6] = 0
    DISTRIBUTION[random_6] += 1

print DISTRIBUTION

DISTRIBUTION_M = dict()

for index in range(10000):
    random_m = get_rand_m(22)
    if random_m not in DISTRIBUTION_M.keys():
        DISTRIBUTION_M[random_m] = 0
    DISTRIBUTION_M[random_m] += 1

print DISTRIBUTION_M
```
